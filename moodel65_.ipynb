```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import cv2 

emotions = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
classes = len(emotions)
print(classes)
```

```python
import pandas as pd
import numpy as np
import cv2

with tpu_strategy.scope():
#     data = pd.read_csv(r'/kaggle/input/data1.csv',names = ['image','image2'])
   label = pd.read_csv(r'../input/fe2013landmarks/labeldata1.csv',names = ['label'])
```

```python
train_set_x_orig=[]
with tpu_strategy.scope():
    for id in range(24942): 
        image = cv2.imread('../input/ldzipp/landmarks2/'+str(id)+'.png') 
        image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image,(48,48))
        array = np.array(image)
        print(array.shape)
        train_set_x_orig.append(array)
```

```python
train_set_x_orig = np.array(train_set_x_orig)
print(train_set_x_orig.shape)
```

```python
train_set_x_orig = train_set_x_orig.reshape((24942,48,48,1))
print(train_set_x_orig.shape)
```

```python
train_set_y_orig = np.array(label.label).astype(int)
print(train_set_y_orig.shape[0])
print(train_set_y_orig)
```

```python
from sklearn.model_selection import train_test_split
train_set_x_orig, test_set_x_orig, train_set_y_orig, test_set_y_orig = train_test_split(train_set_x_orig, train_set_y_orig,shuffle = True, stratify = train_set_y_orig, random_state = 2020, test_size = 0.2 )
print(train_set_x_orig.shape)
print(test_set_x_orig.shape)
print(test_set_x_orig.shape)
print(test_set_y_orig.shape)
```

```python
# Normalize image vectors
X_train = train_set_x_orig/255
X_test = test_set_x_orig/255

# Reshape
Y_train = train_set_y_orig.T
Y_test = test_set_y_orig.T


print ("number of training examples = " + str(X_train.shape[0]))
print ("number of test examples = " + str(X_test.shape[0]))
print ("X_train shape: " + str(X_train.shape))
print ("Y_train shape: " + str(Y_train.shape))
print ("X_test shape: " + str(X_test.shape))
print ("Y_test shape: " + str(Y_test.shape))
```

```python
from keras.utils import to_categorical
Y_train = to_categorical(Y_train, num_classes = classes)
Y_test = to_categorical(Y_test, num_classes = classes)
```

```python
print(Y_train.shape)
print(Y_test.shape)
```

```python
from keras import regularizers
with tpu_strategy.scope():
    def Moodel(input_shape):
   
    
        X_input = Input(input_shape)
    
    #Block 1 -- conv2d - MaxPooling
    
        X = ZeroPadding2D((3,3))(X_input)
    
        X = Conv2D(32,(3,3), strides = (1,1), name = 'conv0')(X)
        X = BatchNormalization(axis=3, name = 'bn0')(X)
        X = Activation('relu')(X)
    
    #MaxPool
    
        X = MaxPooling2D((3,3), strides = (1,1), name = 'max_pool_0')(X)
    
    
    #Block2 -- conv2D -- maxpool2D
    
        X = Conv2D(64,(3,3), name = 'conv1')(X)
        X = BatchNormalization(axis=3, name = 'bn1')(X)
        X = Activation('relu')(X)
    
    #MaxPool
    
        X= MaxPooling2D((2,2), strides = (2,2), name = 'max_pool_1')(X)
    
    
    #Block3 -- conv2D -- MaxPool2D
    
        X = Conv2D(128,(3,3), name ="conv2")(X)
        X = BatchNormalization(axis = 3, name = 'bn2')(X)
        X = Activation("relu")(X)
    
    #MaxPool
    
        X = MaxPooling2D((3,3), strides = (1,1), name = 'max_pool_2')(X)
    
    
    #Block4 -- conv2D -- MaxPool2D
    
        X = Conv2D(256,(3,3), name ="conv3")(X)
        X = BatchNormalization(axis = 3, name = 'bn3')(X)
        X = Activation("relu")(X)
    
    #MaxPool
    
        X = MaxPooling2D((2,2), strides = (2,2), name = 'max_pool_3')(X)
    
    
    #Block5 -- conv2D -- MaxPool2D
    
        X = Conv2D(512,(3,3), name ="conv4")(X)
        X = BatchNormalization(axis = 3, name = 'bn4')(X)
        X = Activation("relu")(X)
    
    #MaxPool
    
        X = MaxPooling2D((2,2), strides = (2,2), name = 'max_pool_4')(X)
    
    #Block6 --flatten --dense
    
        X = layers.Dropout(0.5)(X)
        X = Flatten()(X)
        X = Dense(1500, activation = 'relu', name = 'fc')(X)
        X = Dense(500, activation = 'relu', name = 'fc1')(X)
        X = Dense(64, activation = 'relu', name = 'fc2')(X)
        X = Dense(7, activation = 'softmax', name = 'fc3')(X)
    
   
    
     #Block5 
    
        model = Model(inputs = X_input, outputs = X, name = 'Moodel')
    
        return model
```

```python
with tpu_strategy.scope():
    moodel = Moodel(X_train.shape[1:])
```

```python
moodel.summary()
```

```python
lr_schedule = keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=2e-1,decay_steps=10000,decay_rate=0.9,)
opt = keras.optimizers.SGD(learning_rate=lr_schedule)
with tpu_strategy.scope():
    moodel.compile(optimizer = opt, loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
print(X_train.shape)
```

```python
my_callback = [tf.keras.callbacks.EarlyStopping(monitor = 'acc', patience = .90)]
history = moodel.fit(X_train,Y_train, epochs = 200, batch_size=200, verbose = 1, validation_split = 0.2, callbacks = my_callback)
```

```python
history = moodel.fit(X_train,Y_train, epochs = 50, batch_size=160, verbose = 1, validation_split = 0.2, callbacks = my_callback)
```

```python

moodel.evaluate(X_test,Y_test)
```
